```python
from mtcnn import MTCNN
import cv2
import os
from tqdm import tqdm
detector=MTCNN()
def extract_faces_from_folder(input_dir,output_dir):
    os.makedirs(output_dir,exist_ok=True)
    folders=os.listdir(input_dir)    
    for folder in tqdm(folders):
        folder_path=os.path.join(input_dir,folder)
        if not os.path.isdir(folder_path):
            continue
        output_person=os.path.join(output_dir,folder)
        os.makedirs(output_person,exist_ok=True)
        for img_file in os.listdir(folder_path):
            img_path=os.path.join(folder_path,img_file)
            img=cv2.imread(img_path)
            if img is None:
                continue
            rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
            faces=detector.detect_faces(rgb)
            if len(faces)==0:
                continue
            x,y,w,h=faces[0]['box']
            face=img[y:y+h,x:x+w]
            save_path=os.path.join(output_person,img_file)
            cv2.imwrite(save_path,face)
extract_faces_from_folder(r"data\frames\real",r"data\faces\real")
extract_faces_from_folder(r"data\frames\fake",r"data\faces\fake")
```
