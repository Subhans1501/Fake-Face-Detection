```python
import os
import shutil
import random
SOURCE_DIR=r"D:\Study\Semester 5\ML-LAB\Fake Face Detection\data\faces"
TARGET_DIR=r"D:\Study\Semester 5\ML-LAB\Fake Face Detection\data\faces_split"
SPLIT={"train":0.7,"val":0.15,"test":0.15}
random.seed(42)
def split_class(class_name):
    class_path=os.path.join(SOURCE_DIR,class_name)
    folders=os.listdir(class_path)
    random.shuffle(folders)
    n=len(folders)
    train_end=int(SPLIT["train"]*n)
    val_end=train_end+int(SPLIT["val"]*n)
    splits={
        "train":folders[:train_end],
        "val":folders[train_end:val_end],
        "test":folders[val_end:]
    }
    for split,split_folders in splits.items():
        for folder in split_folders:
            src=os.path.join(class_path,folder)
            dst=os.path.join(TARGET_DIR,split,class_name,folder)
            os.makedirs(os.path.dirname(dst),exist_ok=True)
            shutil.copytree(src,dst)
for cls in ["real","fake"]:
    split_class(cls)
print("✅ Dataset split completed.")
```
